```python
import subprocess
def run_cmd(cmd):
    try:
        res = subprocess.run(cmd, shell=True, capture_output=True, text=True, cwd="C:\\Users\\GMT\\Downloads\\uthenga")
        print(f"STDOUT:\n{res.stdout}\nSTDERR:\n{res.stderr}")
    except Exception as e:
        print(f"Error: {e}")

run_cmd("git status")
```
